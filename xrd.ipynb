```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
from pymatgen.analysis.diffraction.core import DiffractionPattern
import matplotlib.pyplot as plt
```

```python
with MPRester("TcDpUTe7wLxFMs0t") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-675349")
```

```python
xrdCalc = xrd.XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
Diffraction = xrdCalc.get_pattern(structure,scaled=True,two_theta_range=(0, 90))
```

```python
Diffraction
```

```python
xrdCalc.show_plot(structure)
#ce qui nous intresse,c'est les trois premiers pics, on peut faire abstraction de la fin peu lisible
```

```python
Theta = Diffraction.x 
Intensity = Diffraction.y
```

```python
print('Les indices de Bravais-Miller associs aux trois premiers pics du diffractogramme et leur multiplicits respectives sont:')
print(Diffraction.hkls[0])
print(Diffraction.hkls[1])
print(Diffraction.hkls[2])
print('REMARQUE : Pour des structures  symtrie hexagonale, on peut utiliser une notation  4 indices (hkil) connue sous le nom de notation de Bravais-Miller.')
print('Les 3 premiers indices sont lis par symtrie. ')
print('Le troisime indice i est dit redondant car on sait le calculer  partir des deux premiers indices, par la relation i=-(h+k)')
print('Les indices h,k et l sont donc suffisants pour dfinir un plan  eux seuls.')
print('')
print('')
print('Le premier pic est caractris par:')
print('2Theta =', Theta[0],'')
print('Intensit =', Intensity[0])
print('Indices hkl associs : (0,0,1)')
print('')
print('Le deuxime pic est caractris par:')
print('2Theta =', Theta[1],'')
print('Intensit =', Intensity[1])
print('Indices hkl associs : (1,0,0)')
print('')
print('Le troisime pic est caractris par:')
print('2Theta =', Theta[1],'')
print('Intensit =', Intensity[1])
print('Indices hkl associs : (0,0,2)')
```
