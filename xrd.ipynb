```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
from pymatgen.analysis.diffraction.core import DiffractionPattern
import matplotlib.pyplot as plt
```

```python
with MPRester("TcDpUTe7wLxFMs0t") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-675349")
```

```python
xrdCalc = xrd.XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
Diffraction = xrdCalc.get_pattern(structure,scaled=True,two_theta_range=(0, 90))
```

```python
Diffraction
```

```python
xrdCalc.show_plot(structure)
#ce qui nous intéresse,c'est les trois premiers pics, on peut faire abstraction de la fin peu lisible
```

```python
Theta = Diffraction.x 
Intensity = Diffraction.y
```

```python
print('Les indices de Bravais-Miller associés aux trois premiers pics du diffractogramme et leur multiplicités respectives sont:')
print(Diffraction.hkls[0])
print(Diffraction.hkls[1])
print(Diffraction.hkls[2])
print('REMARQUE : Pour des structures à symétrie hexagonale, on peut utiliser une notation à 4 indices (hkil) connue sous le nom de notation de Bravais-Miller.')
print('Les 3 premiers indices sont liés par symétrie. ')
print('Le troisième indice i est dit redondant car on sait le calculer à partir des deux premiers indices, par la relation i=-(h+k)')
print('Les indices h,k et l sont donc suffisants pour définir un plan à eux seuls.')
print('')
print('')
print('Le premier pic est caractérisé par:')
print('2Theta =', Theta[0],'°')
print('Intensité =', Intensity[0])
print('Indice de miller associé : (0,0,1)')
print('')
print('Le deuxième pic est caractérisé par:')
print('2Theta =', Theta[1],'°')
print('Intensité =', Intensity[1])
print('Indice de miller associé : (1,0,0)')
print('')
print('Le troisième pic est caractérisé par:')
print('2Theta =', Theta[1],'°')
print('Intensité =', Intensity[1])
print('Indice de miller associé : (0,0,2)')
```
